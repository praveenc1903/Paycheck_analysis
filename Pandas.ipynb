```python
df['Made_in'] = df['make'].apply(made_in)
```

```python
def priority(price,color):
  if price >= 35000 and color =='red':
    return 'High profit'
  elif price < 35000 and color=='white':
    return 'Less profit'
  else:
    return 'Moderate profit'
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate distance between two points
def distanceCal(x1, y1, x2, y2):
    return np.sqrt((x2 - x1)**2 + (y2 - y1)**2)

# Points to be plotted
x1, y1 = 1, 2  # First point (x1, y1)
x2, y2 = 4, 6  # Second point (x2, y2)

# Calculate the distance
distance = distanceCal(x1, y1, x2, y2)

# Create the plot
plt.figure(figsize=(16, 16))
plt.plot([x1, x2], [y1, y2], marker='o', color='b', linestyle='-', linewidth=2, markersize=8)

# Annotate the points
plt.text(x1, y1, f'({x1}, {y1})', fontsize=12, ha='right', color='blue')
plt.text(x2, y2, f'({x2}, {y2})', fontsize=12, ha='left', color='red')

# Label the distance
midpoint_x = (x1 + x2) / 2
midpoint_y = (y1 + y2) / 2
plt.text(midpoint_x, midpoint_y, f'Distance = {distance:.2f}', fontsize=12, color='green', ha='center')

# Set the plot limits and labels
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.axhline(1, color='black',linewidth=1)
plt.axvline(0, color='red',linewidth=10)
plt.grid(True)

# Title and show plot
plt.title('Distance Between Two Points')
plt.gca().set_aspect('equal', adjustable='datalim')
plt.show()
```
